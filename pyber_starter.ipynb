```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city= pd.read_csv(city_data_to_load)
ride=pd.read_csv(ride_data_to_load)
```

```python
# Combine the data into a single dataset
city.head()
```

```python
ride.head()
```

```python
com=pd.merge(ride, city , how="outer" ,on="city")
```

```python
# Display the data table for preview
```

```python
city_group=com.groupby("city")

city_group.head(1)
```

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

## Bubble Plot of Ride Sharing Data

```python
# Build the scatter plots for each city types
cc=com["type"].value_counts()
cc
```

```python
urban=com[com["type"]=="Urban"]
suburban=com[com["type"]=="Suburban"]
rural=com[com["type"]=="Rural"]
urban.head()
```

```python
uRideCount=urban.groupby(["city"]).count()["ride_id"]
uAvg=urban.groupby(["city"]).mean()["fare"]
uDriverCount=urban.groupby(["city"]).mean()["driver_count"]

sRideCount=suburban.groupby(["city"]).count()["ride_id"]
sAvg=suburban.groupby(["city"]).mean()["fare"]
sDriverCount=suburban.groupby(["city"]).mean()["driver_count"]

rRideCount=rural.groupby(["city"]).count()["ride_id"]
rAvg=rural.groupby(["city"]).mean()["fare"]
rDriverCount=rural.groupby(["city"]).mean()["driver_count"]
```

```python
plt.scatter(uRideCount, uAvg, s=5*uDriverCount, edgecolor="black",  marker="o",alpha=.8, label="Urban" , facecolors="orange")

plt.scatter(sRideCount, sAvg, s=5*sDriverCount, edgecolor="black",  marker="o", alpha=0.6, label="Suburban" , facecolors="skyblue")

plt.scatter(rRideCount, rAvg, s=5*rDriverCount, edgecolor="black",  marker="o", alpha=0.4, label="Rural" , facecolors="yellow")

plt.title("Pyber Ride Sharing Data (2016)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.xlim((0,41))
plt.ylim((18,45))
plt.grid(True)


lgnd = plt.legend( scatterpoints=1, loc="best", title="City Types")
lgnd.legendHandles[0]._sizes = [50]
lgnd.legendHandles[1]._sizes = [50]
lgnd.legendHandles[2]._sizes = [50]

plt.text(43, 30, 'Note:\nCircle size correlates with driver count per city.')

plt.show()
```

```python
# Save Figure
plt.savefig("../images/homework.png")
```

## Total Fares by City Type

```python
typeFare = com.groupby('type')["fare"].sum().reset_index()
typeFare
```

```python
colors=("orange","skyblue","lightcoral")
total_fares=plt.pie(typeFare["fare"], labels = typeFare["type"], shadow = True, explode = (0,0,0.1) ,colors=colors,   startangle=150, autopct = "%1.1f%%")

plt.title('% of Total Fares by City Type')
plt.show()
```
